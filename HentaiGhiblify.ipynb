```python
import os
import tensorflow as tf
from deepdanbooru import project as dd_project

project_path = 'deepdanbooru_model'
model = dd_project.load_model_from_project(project_path, compile_model=False)
```

```python
import os
from PIL import Image
import numpy as np

def tag_image(image_path, model):
    image = Image.open(image_path).convert('RGB').resize((512, 512))
    image_np = np.array(image) / 255.0
    image_np = image_np.reshape((1,) + image_np.shape)
    predictions = model.predict(image_np)[0]
    with open(f'{project_path}/project_tags.txt', 'r') as f:
        tags = [x.strip() for x in f.readlines()]
    top_tags = [tags[i] for i in np.argsort(predictions)[::-1] if predictions[i] > 0.5]
    return ', '.join(top_tags)

def tag_all_images(folder_path):
    for root, _, files in os.walk(folder_path):
        for filename in files:
            if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.webp')):
                try:
                    full_path = os.path.join(root, filename)
                    tags = tag_image(full_path, model)
                    with open(os.path.splitext(full_path)[0] + ".txt", "w") as f:
                        f.write(tags)
                    print(f"✅ {filename} → {tags}")
                except Exception as e:
                    print(f"❌ Failed: {filename} — {e}")

# 🚀 Launch before the gym
tag_all_images("/content/trainHentaiModels")
```
